```python
#Image compression using K-means from Scratch in Python
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy.io import loadmat
from KMeans import *
```

```python
Image = loadmat("datasets/bird_small.mat")
A = Image["A"]
```

```python
#Checking the shape
A.shape
```

```python
# Display image
img = plt.imshow(A)
```

```python
#Preprocessing image

X = (A/255).reshape(A.shape[0]*A.shape[1],3)
```

```python
def CompressedImage(X, K, num_iter):
    kmeans = KMeans(16)
    initial_centroid = kmeans.kMeansInitCentroids(X)
    compressed_Centroid, conpressed_idx = kmeans.runKmeans(X, initial_centroid, 10)
    X_Compressed = X.copy()
    for i in range(1, K+1):
        X_Compressed[(conpressed_idx == i).ravel(),:] = compressed_Centroid[i-1]
    return X_Compressed
```

```python
X_Compressed = CompressedImage(X, 16, 10)
X_Compressed = X_Compressed.reshape(A.shape[0], A.shape[1], 3)
X = X.reshape(A.shape[0],A.shape[1],3)
```

```python
#Visualize Image

def plot_image(Orginal, Compressed):
    fig, ax = plt.subplots(1, 2)
    Img1 = ax[0].imshow(Orginal)
    Img2 = ax[1].imshow(Compressed)
    for i in range(2):
        title = ['Orginal','Compressed']
        ax[i].set_title(title[i])
```

```python
plot_image(X, X_Compressed)
```
