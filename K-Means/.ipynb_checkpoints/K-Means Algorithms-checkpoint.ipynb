```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from KMeans import *
```

```python
mat = loadmat("datasets/ex7data2.mat")
X = mat['X']
```

```python
plt.plot(X[:, 0], X[:, 1], 'yo')
```

```python
#Cluster Assignment

kmeans = KMeans(3)
idx = kmeans.findClosestCentroids(X, np.array([[3,3],[6,2],[8,5]]))
print('Closest centroids for the first 3 examples:\n', idx[0:3])
```

```python
#Move Centroid

centroids = kmeans.computeCentroids(X, idx)
```

```python
#Visualizing Kmeans

kmeans.plot_KMeans(X, idx, centroids, 10)
```

```python
#Random initialization

initial_centroid = kmeans.kMeansInitCentroids(X)
idx = kmeans.findClosestCentroids(X, initial_centroid)
centroids = kmeans.computeCentroids(X,idx)
kmeans.plot_KMeans(X,idx,centroids,10)
```
